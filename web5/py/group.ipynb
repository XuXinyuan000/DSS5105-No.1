```python
from transformers import pipeline

# 初始化预训练的摘要生成模型
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

# 示例 ESG 数据摘要
esg_summary_text = """
AIA Group in 2023 had an ESG score of 7, with strong environmental and moderate governance performance...
Allianz maintained a strong rating of AA with balanced scores across E, S, and G...
"""

# 生成摘要
summary = summarizer(esg_summary_text, max_length=30, min_length=15, do_sample=False)
print("Automated ESG Summary:", summary[0]['summary_text'])
```
